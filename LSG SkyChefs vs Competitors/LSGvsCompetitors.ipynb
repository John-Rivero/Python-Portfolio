```python
import pandas as pd


file = pd.read_csv("Company Analysis.csv")
```

```python
#SORT REVENUE 
sort_revenue = file.sort_values(by='Revenue', ascending=False)

# Define a custom formatting function
def format_revenue(revenue):
    return '${:,.2f}'.format(revenue)

# Apply the formatting function to the "Revenue" column
sort_revenue['Revenue'] = sort_revenue['Revenue'].apply(format_revenue)

# Display the sorted dataframe with formatted revenue
print(sort_revenue)





```

```python
#SORT EMPLOYEES
sort_employees = file.sort_values(by='Employees', ascending=False)
print(sort_employees)
```

```python
#SORT Founding Date
sort_FoundingDate = file.sort_values(by='Founding Date', ascending=False)
print(sort_FoundingDate)
```

```python
#VISUALIZATION
import numpy as np
import matplotlib.pyplot as plt

# Data
companies = sort_revenue['Company']
revenues = sort_revenue['Revenue'].str.replace('$', '').str.replace(',', '').astype(float)

# Create an array of indices for the x-axis
indices = np.arange(len(companies))

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.barh(indices, revenues, color='skyblue')
plt.yticks(indices, companies)
plt.xlabel('Revenue (in billions USD)')
plt.title('Top Companies by Revenue')

# Add labels to the bars
for i, v in enumerate(revenues):
    plt.text(v + 100, i, '${:,.2f}B'.format(v), va='center', fontsize=10, color='black')

plt.gca().invert_yaxis()  # Invert the y-axis to display the highest revenue at the top
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
companies = sort_employees['Company']
employees = sort_employees['Employees']

# Create an array of indices for the x-axis
indices = np.arange(len(companies))

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.barh(indices, employees, color='skyblue')
plt.yticks(indices, companies)
plt.xlabel('Number of Employees')
plt.title('Top Companies by Number of Employees')

# Add labels to the bars
for i, v in enumerate(employees):
    plt.text(v + 100, i, '{:,}'.format(v), va='center', fontsize=10, color='black')

plt.gca().invert_yaxis()  # Invert the y-axis to display the largest number of employees at the top
plt.tight_layout()
plt.show()
```
